```python
# libraries
import pandas as pd
```

```python
# load data from files
companies = pd.read_csv('../data/companies.csv')
education = pd.read_csv('../data/education.csv')
industries = pd.read_csv('../data/industries.csv')
jobs_detail = pd.read_csv('../data/jobs_detail.csv')
jobs = pd.read_csv('../data/jobs.csv')
skills = pd.read_csv('../data/skills.csv')
```

```python
# consolidated columns printed:
dfs = [companies, education, industries, jobs_detail, jobs, skills]
df_names = ['companies', 'education', 'industries', 'jobs_detail', 'jobs', 'skills']

print('Columns in each df:\n')
for i in range(len(df_names)):
    print(f'--- {df_names[i]} ---')
    print(dfs[i].columns, '\n')
```

```python
# inspect possible foreign keys
print('Jobs -- company_id vs. copmanies -- id')
print(sorted(jobs['company_id'].unique()))
print(sorted(companies['id'].unique()))

print('\nJobs -- Education_id vs. Education -- id')
print(sorted(jobs['education_id'].unique()))
print(sorted(education['id'].unique()))

print('\nJobs -- skills_requirement vs. skills -- id')
print(sorted(jobs['skills_requirement'].unique()))
print(sorted(skills['id'].unique()))

print('\nCompanies -- industry_id vs. Industries -- id')
print(sorted(companies['industry_id'].unique()))
print(sorted(industries['id'].unique()))

print('\nIndustries -- industry_skills vs. skills -- skill')
print(sorted(industries['industry_skills'].unique()))
print(sorted(skills['skill'].unique()))
```

```python
print('Dtypes in each df:\n')
for i in range(len(df_names)):
    print(f'--- {df_names[i]} ---')
    print(dfs[i].dtypes, '\n')
```
